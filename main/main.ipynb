```python
class Node :
    def __init__( self , x , y ) :
        self.x = x
        self.y = y

        self.neighbours = []
        #self.in_nodes = []
        #self.out_nodes = []

    def __str__( self ) :
        return f"{self.x},{self.y}"
    def __repr__(self):
         return f"({self.x},{self.y})"
```
